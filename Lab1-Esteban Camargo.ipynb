## Problema 1

```python
def colores(imagen, color):
    
    #Leemos imagen y convertimos a RGB
    img = cv2.imread(imagen)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    #Definimos alto y ancho de la imagen
    alto = img.shape[0]
    ancho = img.shape[1]
    lienzo = np.zeros((alto,ancho,3))
    
    #Definicion funcion para lectura de pixeles
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i,j]

            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]

            if color == 1:
                lienzo[i,j] = [blue,0,0]
            elif color == 2:
                 lienzo[i,j] = [0,green,0]
            elif color == 3:
                 lienzo[i,j] = [0,0,red]
            elif color == 10:
                 lienzo[i,j] = [0,green,red]
            elif color == 20:
                 lienzo[i,j] = [blue,green,0]
            elif color == 30:
                 lienzo[i,j] = [blue,0,red]
     
    #Creamos imagen
    cv2.imwrite('Problema1/monocromo.jpg', lienzo)
    
    #Leemos imagen para mostrar
    colores = cv2.imread('Problema1/monocromo.jpg')
    colores = cv2.cvtColor(colores, cv2.COLOR_BGR2RGB)
    
    plt.imshow(colores)
    
    return plt.show()
```

## Problema 2

```python
def integracion_imagenes(imagen_1, imagen_2, imagen_3):

    imagenes = pd.Index([imagen_1, imagen_2, imagen_3])
    
    azul =  pd.DataFrame(imagenes[imagenes.str.contains("azul")], columns=["Nombre"])
    verde = pd.DataFrame(imagenes[imagenes.str.contains("verde")], columns=["Nombre"])
    rojo =  pd.DataFrame(imagenes[imagenes.str.contains("rojo")], columns=["Nombre"])
    
    img_azul = cv2.imread(azul.at[0,"Nombre"])
    img_azul = cv2.cvtColor(img_azul, cv2.COLOR_BGR2RGB)

    img_verde = cv2.imread(verde.at[0,"Nombre"])
    img_verde = cv2.cvtColor(img_verde, cv2.COLOR_BGR2RGB)            

    img_rojo = cv2.imread(rojo.at[0,"Nombre"])
    img_rojo = cv2.cvtColor(img_rojo, cv2.COLOR_BGR2RGB)    

    #Definimos alto y ancho de la imagen
    alto = img_azul.shape[0]
    ancho = img_azul.shape[1]
    lienzo = np.zeros((alto,ancho,3))

    for i in range(0, alto):
            for j in range(0, ancho):
                pixel_azul = img_azul[i,j]
                pixel_verde = img_verde[i,j]
                pixel_rojo = img_rojo[i,j]

                blue = pixel_azul[2]
                green = pixel_verde[1]
                red = pixel_rojo[0]

                lienzo[i,j] = [red,green,blue]

    #Creamos imagen
    #nombre = imagen_1.split("/")[0]
    #cv2.imwrite('Problema2/'+nombre+'.jpg', lienzo)
                
    plt.imshow(lienzo.astype('uint8'))
    return plt.show()
```

```python
def imagen_gris_3d(imagen):

    img = cv2.imread(imagen)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    #Definimos alto y ancho de la imagen
    alto = img.shape[0]
    ancho = img.shape[1]
    lienzo_azul = np.zeros((alto,ancho,3))
    lienzo_verde = np.zeros((alto,ancho,3))
    lienzo_rojo = np.zeros((alto,ancho,3))

    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i,j]

            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]

            lienzo_azul[i,j] = [blue,blue,blue]
            lienzo_verde[i,j] = [green,green,green]
            lienzo_rojo[i,j] = [red,red,red]

    #Creamos imagen
    cv2.imwrite('Problema3/grises/montaas_azul.jpg', lienzo_azul)
    cv2.imwrite('Problema3/grises/montaas_verde.jpg', lienzo_verde)
    cv2.imwrite('Problema3/grises/montaas_rojo.jpg', lienzo_rojo)

    return print("Imagenes creadas")
```

## Problema 4

```python
#Imagen po utilizar
img = cv2.imread('Problema4/montaas.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img.astype('uint8'))
plt.show()
```

```python
def histogramas(imagen):
    
    img = cv2.imread(imagen)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    #Definimos alto y ancho de la imagen
    alto = img.shape[0]
    ancho = img.shape[1]

    #lienzos
    gris_ponderado = np.zeros((alto,ancho,1))
    frecuencias = np.zeros((256))
    
    #Calculamos el gris ponderado
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i,j]
            frecuencias[pixel] = frecuencias[pixel] + 1
            
            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]

            gris_ponderado[i,j] = int(0.299*blue + 0.587*green + 0.114*red)
            gris_ponderado = np.float32(gris_ponderado)

    #Calculamos densidad
    densidad = np.true_divide(frecuencias,(alto*ancho)) #calculamos densidad
    
    #CDF
    cdfDensidad = np.cumsum(densidad, dtype = float)
    mean_cdf = cdfDensidad.mean()*100

    #Realizamos los histogramas
    f = plt.figure()
    f.set_figwidth(20)
    f.set_figheight(10)

    #Histogramas de color
    colors = ('b','g','r')
    for i,color in enumerate(colors):
        hist = cv2.calcHist([img],[i],None,[256],[0,256])
        plt.plot(hist,color = color)

    #Histograma gris ponderado
    hist1 = cv2.calcHist([gris_ponderado],[0],None,[256],[0,256])
    plt.plot(hist1, color= "gray")
    
    #Promedio distribucin
    plt.axvline(x = mean_cdf, color = 'r', linestyle = '-')

    plt.xlabel("Intensidad de Pixel")
    plt.ylabel("Densidad")
    plt.title("Comparacin de histogramas")

    return plt.show()
```

```python
histogramas("Problema4/montaas.jpeg")
```

## Problema 6

Que es el espacio de color HSV?

El espacio de color HSV, segn como lo denotan sus siglas, define los colores en 3 distintos factores que son los siguientes.

1. Matiz: se refiere basicamente al color del espacio que se esta observando.Este est expresado por un nmero del 0 a 360. Los colores se definen de la siguiente forma:
    - Rojo: 0 - 60 grados
    - Amarillo: 61-120 grados
    - Verde: 121-180 grados
    - Cyan: 181-240 grados
    - Azul: 241- 300 grados
    - Magenta: 301-360 grados


2. Saturacion: la saturacin, en el modelo HSV, define la cantidad de gris en un color en particular y puede tomar valores de 0 a 100 por ciento. El efecto de reducir este componente haca 0, hace que la imagen sea ms gris y produce un efecto de sombra. Adicionalmente, el rango de este componente tambin puede ser expresado como 0 a 1; en donde 0 es gris y 1 es algn color.


3. Valor/Brillo: El valor o brillo funciona como complemento a la saturacin ya que se define a la intensidad o el brillo que toma el color. Este componente puede tomar valores del 0-100% o del 0-1, al igual que la saturacin. Mientras el valor se acerce ms a 0, ms negro ser mientras que 100/1 ser el color ms brillante y se podr observar mejor el color.


Algunos de los usos del espacio HSV es cuando los diseadores seleccionan algn color en especfico para pintura o tinta debido a que este espacio reprensenta de una mejor manera como las personas observar y se relacionan con el color, que las otras escalas conocidas.on los colores que el modelo de color RGB. Otra utilidad de este espacio es que contribuye mucho ms a los grficos de alta calidad y es por esto que puede ser encontrado en muchos programas de software de edicin de imagenes de alta gama, videojuegos, tarjetas grficas, etc.
